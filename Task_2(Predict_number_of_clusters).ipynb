```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Try different number of clusters from 1 to 10
num_clusters = range(1, 11)
inertias = []
X = df.iloc[:, [1, 2, 3, 4]].values
for k in num_clusters:
    # Fit KMeans clustering algorithm to the data
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    
    # Compute inertia (sum of squared distances of samples to their closest cluster center)
    inertia = kmeans.inertia_
    inertias.append(inertia)

# Plot the elbow curve to determine the optimal number of clusters
plt.plot(num_clusters, inertias, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve for Iris Dataset')
plt.show()
```

# CONCLUSION

# The elbow curve shows the relationship between the number of clusters and the inertia. The "elbow" point in the curve represents the optimal number of clusters, where adding more clusters does not lead to a significant reduction in inertia. In this case, we can see that the elbow point is at k=3, so we can conclude that the iris dataset has 3 clusters.

# Visualizing the clusters

```python
# Fit KMeans clustering algorithm to the data
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Get the predicted cluster labels
labels = kmeans.predict(X)

# Visualize the clusters using the first two features (Sepal Length and Sepal Width)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('KMeans Clustering on Iris Dataset')
plt.show()
```

# This visualization produces a scatter plot of the first two features (Sepal Length and Sepal Width) of the iris dataset, where each point is colored according to its predicted cluster label. Since we determined that the optimal number of clusters is 3, we set n_clusters=3 in the KMeans algorithm. # 3D PLOT FOR VISUALIZING ALL FOUR FEATURES

```python
from mpl_toolkits.mplot3d import Axes3D

# Fit KMeans clustering algorithm to the data
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Get the predicted cluster labels
labels = kmeans.predict(X)

# Visualize the clusters using all four features (Sepal Length, Sepal Width, Petal Length, Petal Width)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=labels)
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
plt.title('KMeans Clustering on Iris Dataset')
plt.show()
```
